## Using `requests`
The `requests` library lets you make HTTP requests to web sites just like your browser does. There are different types of HTTP requests, but the only ones that we'll probably ever worry about are `GET` requests and `POST` requests. Today, we'll only be looking at `GET` requests. When you type some web address in your browser and press `Enter`, the browser executes a `GET` request and handles the website's response for you. We can do this ourselves, without the browser, using the `requests` library. 

```python
# from IPython.display import HTML
# HTML(requests.get("http://www.google.com").content)
```

```python
# Let's try to hit StreetEasy's rentals/data endpoint
resp = requests.get("http://streeteasy.com/nyc/api/rentals/data", params={"key": apikey})
resp
```

__PRO TIP:__ If you have some Python object (like `resp`, above) and you want to know more about it, try using `resp.__dict__` or `dir(resp)`. Accessing an object's `__dict__` attribute will tell you all about the values of all of all of that object's attributes. `dir()` will tell you what methods (functions) and attributes (values) are associated with the object.

```python
# Let's get some info
print "Ok?", resp.ok
print 
print "Status:", resp.status_code
print 
print "URL:", resp.url
print 
print "Cookies:", resp.cookies
print 
print "Headers:", resp.headers
```

```python
# And the motherlode
resp.content
```

The part of the response that we're worried about is in the `.content` attribute. That's where all of the good stuff is. The `.content` field is a string, and in this case, it is a string with JSON format. It's not particularly useful to us as a string so let's turn it into a dictionary. 

```python
import json
# json.loads(s) takes a JSON string, s, and converts it into Python data structures.
# Check out help(json) for more info. 
example = '{"example_key": "example_value"}'
json.loads(example)['example_key']
```

```python
# our response object, `resp`, actually has a .json() method, which
# does the equivalent of json.loads(resp.content)
data = resp.json()

# Now that `data` is a dictionary, we can access specific fields
# and perform computations like
average_square_yards = data["average_sqft"] / (3**2)
print "Average Square Yardage:", average_square_yards
```

```python
# Let's try another endpoint
search_url = "http://streeteasy.com/nyc/api/sales/search"
search_resp = requests.get(search_url, params={"key": apikey})
sales_data = search_resp.json()
```
