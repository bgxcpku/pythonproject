# Web Scraping With Requests and BeautifulSoup
For this part, we're going to try to scrape some data from Airbnb by making a series of HTTP requests using the `requests` library and parsing the HTML responses using `BeautifulSoup`, which will allow us to access the parts of the HTML that we're interested in more easily than would basic string parsing. 

````python
import requests
from BeautifulSoup import BeautifulSoup
import re  # Regular expressions
````

After poking around the Airbnb website and making a few searches, it became clear that you can get listings in a particular city by creating a URL of the format below. For each city site/URL, there are a number of pages of listings (there are 56 pages of Boston results), and each listing has its own id and corresponding link/site/URL. In order to collect data for all of the listings for a particular city, we'll have to visit the individual listing url (e.g https://www.airbnb.com/rooms/4359160) for each listing on each page of the city results. I'll walk through an example of how we might do this for one particular listing in one particular city, and then we can think about how to generalize this so that we can gather and store data for oodles of listings in oodles of cities. 

````python
# The base URL. The {city} bit can be replaced with whatever 
# city we want by using .format 
url = "http://www.airbnb.com/s/{city}"
````

````python
# Make an initial request to "http://www.airbnb.com/s/Boston" 
resp = requests.get(url.format(city="Boston"))
````

````python
# Here we initialize a BeautifulSoup object from the HTML content of the response. 
# Remember, `resp.content` is a string with the HTML source. 
soup = BeautifulSoup(resp.content)

# Uncomment this to see the parsed HTML string
# soup
````

This page has a bunch of results for the city of Boston. I want to find all the links that link to pages for individual listings on this page. By poking around in dev tools on the web site, I found that all of the URLs that link to listing pages are in anchor tags (expected, as all links are), but they also have `href` attributes that start with `/rooms/<some number>`. What I'd like to do is find all the anchor tags that have an href that matches this pattern. 

`BeautifulSoup` has a `.findAll` method, which will help us. It allows us to match on tag type (name) as well as other attributes of the tag, and it happens that `href` is an attribute of the `<a>` tag.

````python
help(soup.findAll)
````

It looks like we want to pass this method the string, `"a"` for the name argument, and then for the `attrs` argument, we'll have to pass a dictionary with key `"href"` and value of some function that returns `True` when the `href` matches our pattern and `False` otherwise. Let's define such a function.

````python
def is_room_link(href):
    """
    Takes a string, href, and returns True if the string 
    starts with "/rooms". Otherwise, it returns False. 
    """
    # This checks that href exists (i.e. it isn't None or an empty string)
    # and that it starts with "/rooms", which is what we want. I only 
    # added the `if href` check because it was breaking without it. This 
    # was happening because not every anchor tag has an href. 
    if href and href.startswith("/rooms"):
        return True
    else:
        return False

# A quick test
print is_room_link("/rooms/3029802"), is_room_link("/not/a/room/link")
````

````python
# Let's find our links!
links = soup.findAll(name="a", attrs={"href": is_room_link})
links
````

````python
# That's a whole lot of crap! Let's take a look at the first link
a = links[0]
a.attrs
````

````python
# We just want the hrefs!
addrs = [a['href'] for a in links]
addrs[:10]
````

````python
# We also don't want those "/rooms/new" links, since they're for us to list our own space, which we don't want to do. 
# Let's get rid of those. 
addrs = [addr for addr in addrs if addr != "/rooms/new"]
addrs[:10]
````

````python
# Yay. These URLs are "relative" to the host, which is "http://www.airbnb.com". 
# To get the whole URL, add the host bit of the url to the beginning of each string
room_urls = ["http://www.airbnb.com" + addr for addr in addrs]
room_urls[0]
````

````python
# If you copy and paste that link in your browser, it'll take you to an 
# individual listing page, which has a bunch of info about the particular listing. 
room_page = requests.get(room_urls[0])
room_soup = BeautifulSoup(room_page.content)
````

Let's find the following information about each listing:

- Price
- Location (Latitude, Longitude, Address)
- Number of Beds
- A list of amenities
- Rating

One nice thing that I found in inspecting the source code of this page is that Airbnb includes some summary info in some `<meta>` tags inside the `<head>` of the HTML. See: 

```
<meta property="og:description" content="Apartment in Somerville, United States. Well located room with a single bed. Close to Boston and Cambridge, a 12 minute walk or 3 minute bus ride to the subway that takes you downtown in 12 minutes, a 3 minute walk to a bus that takes you to Harvard Square in 12 minutes  (red line) and ...">
      <meta property="airbedandbreakfast:locality" content="Somerville">
      <meta property="airbedandbreakfast:region" content="MA">
      <meta property="airbedandbreakfast:country" content="United States">
      <meta property="airbedandbreakfast:city" content="Somerville">
      <meta property="airbedandbreakfast:rating" content="5.0">
      <meta property="airbedandbreakfast:location:latitude" content="42.38152001216975">
      <meta property="airbedandbreakfast:location:longitude" content="-71.08619692421473">
      ```

````python
# Some fancy stuff because it's faster to type and I'm tired :)
meta_props = room_soup.findAll("meta", attrs={"property": lambda x: x.startswith("airbedandbreakfast") if x else False})
geo_info = {p['property'].split(":")[-1]: p["content"] for p in meta_props}
geo_info
````

Theres some nifty info! Now let's get some amenities!

````python
amenities_div = room_soup.findAll("div", attrs={"class": "row-space-1"})
amenities_div
````

````python
# Let's get a list of the amenities as simple strings. Based on the above, we want to 
# end up with something like this: ["Internet", "Free Parking on Premises", "Wireless Internet", "Family/Kid Friendly"]
amenities = []
for sub_soup in amenities_div:
    # Each element in the list above has one `span` tag within it
    # The text content of that span element is what we're after. 
    
    # Get the span. Remember that .findAll returns a list, 
    # so we need to get the first element (there should only be one)
    span = sub_soup.findAll("span")[0]
    span_text = span.text
    amenities.append(span_text)
    
amenities
````

Hooray! we got our list of amenities. 

````python
info = room_soup.findAll("div", {"class": "col-md-6"})
info[0]
````

````python
properties = []
for sub_div in info:
    for sub_sub_div in sub_div.findAll("div"):
        print sub_sub_div.text
        properties.append(sub_sub_div.text)
````

````python
p_dict = {}
for p in properties:
    if ":" in p:
        head, _, tail = p.partition(":")
        p_dict[head] = tail
p_dict
````

````python
# Now let's get the price!
# I found this in the source: <meta content="35" itemprop='price'>
# which is nice. Let's, use that

price = room_soup.findAll("meta", {"itemprop": "price"})[0]["content"]
price
````

````python
# Let's put it all together
apt_info = {}
apt_info.update(p_dict)  # Add properties
apt_info.update(geo_info)
apt_info["amenities"] = amenities
apt_info["price"] = price
````
